```python
#CALIBRADOR DE COLORES Y COORDENADAS DE LECTURA DE PANTALLA
import pyautogui
import time
from IPython.display import clear_output
def get_color_at_mouse():
    try:
        while True:
            x, y = pyautogui.position()
            pixel_color = pyautogui.screenshot().getpixel((x, y))
            print(f"Mouse: (X={x}, Y={y}) - Color: RGB{pixel_color}   ",end='\r')
            time.sleep(0.2)
    except KeyboardInterrupt:
        clear_output()
get_color_at_mouse()
```

```python
#/////////////////////////////////////////////
#///////////////-CONFIGURACION-///////////////
#/////////////////////////////////////////////
#----------------VOY DESPUES DE---------------
palabra = "uriel"
#palabra = "alejandro.l"
#palabra = "uri.gomez"
#palabra = "manuel"
#palabra = "carolina"
#palabra = "alisson.apaiz"
#palabra = "sheyla.araujo"
#palabra = "mariana.licona"

#///////-COORDENADAS-///////
#---------PANTALLA----------
x = 250
y = 750

#*Region_texto y Region_color son funciones que usan los parametros de (x, y, ancho, alto), si ajustas tu ventana a muy peque√±a,
#puedes ajustar el ancho y alto para que no tome colores de fuera de la ventana, podria dar error si, por ejemplo,
#en youtube aparece ese color. No creo que sea necesario modificar el ancho y alto, solo x y y.

#-----------MOUSE-----------
x1 = 340
y1 = 1130

#------CAMBIAR SONIDITO------
Frecuencia = 1000
Duracion = 700

#/////////////////////////////////////////////////////////////
#//////////////-INICIO DE CODIGO NO MODIFICAR-////////////////
#/////////////////////////////////////////////////////////////
import winsound
import pyautogui
import time
import numpy as np
from PIL import Image
import keyboard
import pytesseract
pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"

Turno = False
Color = False
clickconter = 0
contador = 0
region_texto = (x, y, 300, 200) #---Texto---
region_color = (x, y, 150, 350) #---Color---
color = (219, 75, 54)
red_circle = "large"

def find_word(palabra, region):
    global Turno, Color
    screenshot = pyautogui.screenshot(region=region).convert("RGB")
    texto_extraido = pytesseract.image_to_string(screenshot)
    if palabra.lower() in texto_extraido.lower():
        Turno = True

def find_color_in_region(color, region, red_circle):
    global Color
    screenshot = pyautogui.screenshot(region=region).convert("RGB")
    texto_extraido = pytesseract.image_to_string(screenshot)
    screenshot_np = np.array(screenshot)
    target_color = np.array(color)
    result = np.any(np.all(screenshot_np[:,:,:3] == target_color,axis=2))
    if result:
        Color = True
    if red_circle.lower() in texto_extraido.lower():
        Color = True

try:
    while True:
        contador += 1
        position = find_color_in_region(color, region_color, red_circle) #Color
        print(f"Nombre: {Turno}, Color: {Color}, Intento {contador}   ", end='\r')
        if Color == True:
            turno = find_word(palabra, region_texto) #Nombre
            print(f"Nombre: {Turno}, Color: {Color}, Intento {contador}   ", end='\r')
            if Turno == True:
                winsound.Beep(Frecuencia, Duracion)

                if clickconter == 0:
                    #pyautogui.press('space')
                    pyautogui.moveTo(x1, y1)
                    pyautogui.leftClick()
                    clickconter += 1
                Color = False
                winsound.Beep(Frecuencia, Duracion)
            Turno = False
        else:
            clickconter = 0
        Turno = False
        Color = False

        if keyboard.is_pressed('F1'):
            print("F1 presionado, saliendo del programa.")
            break
except KeyboardInterrupt:
    print("\nPrograma detenido por el usuario.")
```
